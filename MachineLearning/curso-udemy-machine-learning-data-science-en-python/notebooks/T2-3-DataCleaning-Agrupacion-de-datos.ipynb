## Agregacion de datos por Categoria.

```python
import numpy as np
import pandas as pd
```

```python
# Agregacion de datos por categoria.
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
# Generacion de 500 datos aleatorios, tomando como referencia las variables gender e income.
n = 500
gender_data = []
income_data = []

for i in range(0,500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
# N (m, s) -> m + s * z
height = 160 + 30 * np.random.rand(n)
wheight = 65 + 25 * np.random.rand(n)
age = 30 + 12 * np.random.rand(n)
income = 18000 + 3500 * np.random.rand(n)
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : wheight,
        "Age" : age,
        "Income" : income
    }
)
```

```python
data.head()
```

```python
## Agrupando datos por medio de variables categoricas.
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
# Imprimiendo por secciones de names y groups.
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
# Obteniendo una seccion del grupo.
Female = grouped_gender.get_group("Female")
Female
```

```python
# Agrupando por mas de una categoria.
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

### Operaciones sobre datos agrupados.

```python
double_group.sum()
```

```python
double_group.mean()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
# calculando valores con aggregation.
double_group.aggregate(
    {
        "Income" : np.sum,
        "Age" : np.mean,
        "Height" : np.std
    }
)
```

```python
# Calculando los valores con funciones lambda.
double_group.aggregate(
    {
        "Age" : np.mean,
        "Height" : lambda h:np.mean(h)/np.std(h)
    }
)
```

```python
# Calculando valores pasando por argumentos las funciones.
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
# Calculando valores pasando la funcion lambda por argumento.
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```
