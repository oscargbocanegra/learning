### Transformaciones entre columnas.

```python
# Convirtiendo altura en metros.

olympics['Height'] / 100
```

```python
# Generando el Indice de Masa corporal en una nueva columna BMI.
olympics['BMI'] = olympics['Weight'] / (olympics['Height'] / 100) ** 2
```

```python
olympics['BMI'].describe()
```

```python
# Consultando los atletas con indice de masa corporal BMI mayor a 60.
olympics[olympics['BMI'] > 60]
```

```python
sb.histplot(olympics['BMI'], bins=20)
```

```python
# Agregando columna Full Name sumando columnas.
olympics['Olympic'] = olympics['City'] + ' ' + olympics['Year'].astype('str')
```

```python
olympics.head()
```

```python
# Modificando valores existentes en las columnas con el metodo MAP.
olympics['Sex_numeric'] = olympics['Sex'].map({'M':1, 'F':0})
```

```python
olympics.loc[:10,['Sex','Sex_numeric']]
```

```python
# Creando nueva columna que valida si el atleta era mayor de edad.
def es_adulto(edad):
    if edad >= 18:
        return 1
    else:
        return 0
```

```python
# Creando nueva columna que valida si el atleta era mayor de edad. con la funcion creada adulto.
olympics['Adult'] = olympics['Age'].apply(es_adulto)
```

```python
olympics.head()
```

```python
olympics.query('Age < 19').head()
```

```python
# aplicando la funcion lambda para calcular altura en metros.
olympics['Height_m'] = olympics['Height'].apply(lambda x: x / 100)
```

```python
# Eliminando atletas duplicados para un evento en particular.
medals = medals.drop_duplicates(subset=['Games', 'Event'])
medals.shape
```

```python
# Creando dataframe de un grupo head()
medals_subset = medals.head()
```

```python
# Agrupando datos para  obtener la cantidad de medallas.
medals_subset.groupby('NOC').count()
```

```python
# Seleccionando la columna deseada.
medals_subset.groupby('NOC').count()['Medal']
```

```python
# Ordenando por medallas acorde con la cantidad de medallas ganadas por pais.
medals_count = medals.get(['NOC', 'Medal']).groupby('NOC', as_index=False).count()
medals_count.head()
```

```python
medals_count.sort_values(by= 'Medal', ascending=False).head()
```

```python
# realizando filtros bor deportes.
olympics_gb = olympics.get(['Sport', 'Age', 'Height','Weight'])
olympics_gb.head()
```

```python
olympics_gb = olympics_gb[olympics_gb['Sport'].isin(['Basketball','Football'])]
olympics_gb.head()
```

```python
# Realizando agrupaciones con opciones de agregacion para cada campo.
olympics_gb.groupby('Sport').agg({'Age': 'mean', 'Height': ['min', 'max']})
```

```python
# Contar los paises con el mayor numero de medallas.
medals_count = medals.get(['NOC', 'Medal']).groupby('NOC', as_index=False).count()
medals_count.head()
```

```python
# Contando por grupo 
medals_count.sort_values(by='Medal', ascending=False).head(10)
```

```python
# Cuentas de medallas por tipo (Oro, plata, bronce) y Pais.
medals.query('Medal == "Gold"')\
    .get(['NOC', 'Medal'])\
    .groupby('NOC', as_index=False)\
    .count()\
    .sort_values(by = 'Medal', ascending = False)\
    .head(10)
```

```python
# Cuentas de medallas por tipo (Oro, plata, bronce) y Pais.
medals.get(['NOC', 'Medal', 'Event'])\
    .groupby(['NOC', 'Medal'], as_index=False)\
    .count()
```

```python
# Realizando  busqueda con el comando pivot_table.
medals.get(['NOC', 'Medal', 'Event']).pivot_table(index = 'NOC', columns = 'Medal', aggfunc = 'count')
```

```python
# Realizando  busqueda con el comando pivot_table filtrando los valores NAN por ceros '0'.
medals_pivot = medals.get(['NOC', 'Medal', 'Event']).pivot_table(index = 'NOC', columns = 'Medal', aggfunc = 'count', fill_value =0)
medals_pivot
```

```python
# se resetea el indice.
medals_pivot = medals_pivot.reset_index()
medals_pivot
```

```python
# Modificando el orden de las columnas..
medals_pivot.columns = ['Gold', 'Silver', 'Bronze']
medals_pivot
```

```python
# Modificando el orden de las columnas.
medals_pivot = medals_pivot[['Gold', 'Silver', 'Bronze']]
medals_pivot
```

```python
# Ordenando por la medalla de ORO.
medals_pivot.sort_values(by = ['Gold'], ascending = False)
```

```python
# Realizando consulta con meals pivotando con la columna NOC.
medals_melt = medals_pivot.melt(id_vars = 'NOC', value_vars = ['Bronze', 'Gold', 'Silver'], var_name = 'Medal', value_name = 'Count')
medals_melt
```

```python
# Realizando filtro con el pais argentina. agrupando las columnas con melt.
medals_melt.query('NOC == "ARG"')
```
