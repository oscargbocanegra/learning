```python
# Importando librrerias
import os
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Consultando arhivos locales en ruta datasets / inputs
path_input_datasets = "../datasets/input/olimpiadas/"
path_output = "../datasets/output/"
os.listdir(path_input_datasets)
#os.mkdir("../datasets/output")
```

```python
olympics = pd.read_csv(path_input_datasets+'athlete_events.csv')
olympics.head()
```

### Realizando Filtros mascaras, query
### Ejemplo para atletas olimpicos peruanos.

```python
olympics[olympics['NOC'] == 'PER']
```

```python
olympics.query('NOC == "PER"')
```

```python
# Realizando filtros conjuntos, Ejemplo los medallistas de Seul 88.
olympics[(olympics['NOC'] == 'PER') & 
         (olympics['Sport'] == 'Volleyball') & 
         (olympics['Year'] == 1988) & 
         (olympics['Medal'] == 'Silver')]
```

```python
# Realizando filtros conjuntos, Ejemplo los medallistas de Seul 88 con la funcion query.
olympics.query('(NOC == "PER") & (Sport == "Volleyball") & (Year == 1988) & (Medal == "Silver")')
```
