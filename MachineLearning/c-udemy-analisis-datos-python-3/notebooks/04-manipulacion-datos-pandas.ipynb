```python
# Importando librrerias
import os
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Consultando arhivos locales en ruta datasets / inputs
path_input_datasets = "../datasets/input/olimpiadas/"
path_output = "../datasets/output/"
os.listdir(path_input_datasets)
#os.mkdir("../datasets/output")
```

```python
olympics = pd.read_csv(path_input_datasets+'athlete_events.csv')
olympics.head()
```

### Realizando Filtros mascaras, query
### Ejemplo para atletas olimpicos peruanos.

```python
olympics[olympics['NOC'] == 'PER']
```

```python
olympics.query('NOC == "PER"')
```

```python
# Realizando filtros conjuntos, Ejemplo los medallistas de Seul 88.
olympics[(olympics['NOC'] == 'PER') & 
         (olympics['Sport'] == 'Volleyball') & 
         (olympics['Year'] == 1988) & 
         (olympics['Medal'] == 'Silver')]
```

```python
# Realizando filtros conjuntos, Ejemplo los medallistas de Seul 88 con la funcion query.
olympics.query('(NOC == "PER") & (Sport == "Volleyball") & (Year == 1988) & (Medal == "Silver")')
```

```python
# Filtros con comparaciones.
olympics[olympics['Age'] < 20].head()
```

```python
# Filtro comparando campos peso mayor que la altura.
olympics.query('Weight > Height')
```

```python
# Filtros con listas.
olympics['Sport'].unique()
```

```python
deportes = ['Tennis','Table Tennis']
```

```python
olympics[olympics['Sport'].isin(deportes)]
```

```python
olympics[olympics['Sport'].isin(deportes) & (olympics['City'] == 'Sydney')]
```

```python
# Filtros isnull 
olympics[olympics['Medal'].isnull()]
```

```python
# Filtros notnull 
# opcion 1 negando la condicion con el comando ~
# olympics[ ~ olympics['Medal'].isnull()]
# opcion 2 con el comando pandas notnull
# Filtros isnull 
olympics[olympics['Medal'].notnull()]
```

```python
# Realizando filtros Negativos.
olympics[olympics['NOC'] != 'CHN']
```

```python
# Realizando filtros negativos con query.
olympics.query('NOC != "CHN"')
```

```python
# Realizando filtros con String.
olympics_rou = olympics[olympics['NOC'] == 'ROU']
```

```python
# Realizando filtro por el innicio del nombre de la atleta Nadia.
olympics_rou[olympics_rou['Name'].str.startswith('Nadia')]
```
