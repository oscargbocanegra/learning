## Ejercicio ordenando archivos en un solo dataset.

```python
import os
import pandas as pd
```

```python
# Consultando arhivos locales en ruta datasets / inputs
path_input_lego = "../datasets/input/lego-database/sets-year/"
path_output = "../datasets/output/"
os.listdir(path_input_lego)
#os.mkdir("../datasets/output")
```

```python
# Ordenando listado.
files = sorted(os.listdir(path_input_lego))
# Concatenando datasets lego con la funcion append.
df_lego_sets = pd.DataFrame()

for file in files:
    lego_year = pd.read_csv(path_input_lego+file)
    df_lego_sets = df_lego_sets._append(lego_year, ignore_index = True)
```

```python
df_lego_sets
```

```python
# Concatenando datasets lego con la funcion concat.
files = sorted(os.listdir(path_input_lego))
dfs = [pd.read_csv(path_input_lego+file) for file in files]
# Concatenando datasets lego con la funcion concat.
df_lego_concat = pd.concat(dfs, ignore_index=True)
```

```python
df_lego_concat
```

```python
# Exportando el dataset final al disco local.
df_lego_concat.to_csv(path_output+'lego_sets.csv', index=False)
```

## 02 Marging DataFrames

```python
# Consultando arhivos locales en ruta datasets / inputs
path_input_lego = "../datasets/input/lego-database/"
path_output = "../datasets/output/"
#os.listdir(path_input_lego)
```

```python
sets = pd.read_csv(path_input_lego+'sets.csv')
sets.head()
```

```python
themes = pd.read_csv(path_input_lego+'themes.csv')
themes.head()
```

```python
# Realizando merge de 2 tablas por los respectivos identificadores.
df_merge = pd.merge(sets,themes, left_on='theme_id', right_on='id')
```

```python
df_merge
```

```python
# Asignando dataframe resultado, 
df_set_with_themes = pd.merge(sets,themes, left_on='theme_id', right_on='id')
df_set_with_themes.head(5)
```

```python
# Eliminando columnas al dataframe creado con merge.
df_set_with_themes = df_set_with_themes.drop('id', axis=1)
df_set_with_themes.head(5)
```

```python
# Renombrando columna al dataframe themes para igualiar el nombre al dataframe set.
themes = themes.rename(columns={'id': 'theme_id'})
themes.head(5)
```

```python
# Realizando merge con set y themes, con el cambio de nombre en la columna themes.id
df_set_with_themes = pd.merge(sets,themes, on = 'theme_id')
df_set_with_themes.head(5)
```

```python
# Renombrando columas acorde con los nombres de las tablas.
df_set_with_themes = pd.merge(sets,themes, on = 'theme_id', suffixes=['_set', '_theme'])
df_set_with_themes.head(5)
```

```python
# Excluyendo columnas que no son necesarias en los cruces.
df_set_with_themes = pd.merge(sets,themes.get(['theme_id','name']), on = 'theme_id', suffixes=['_set', '_theme'])
df_set_with_themes.head(5)
```

### Resolviendo problemas con llaves repetidas.

```python
# Realizando consulta con el metodo query para validar informacion.
sets_example = sets.query("set_num == '7241-1'")
sets_example
```

```python
sets_example = sets.query("set_num == '7241-1'")
sets_example
```

```python
inventory_example = inventory_sets.query("set_num == '7241-1'")
inventory_example
```

```python
# merge de la columna.
pd.merge(sets_example, inventory_example, on = 'set_num')
```

```python
# Merge group by con columnas repetidas.
inventory_sets.get(['set_num','quantity']).head()
```

```python
# Realilzando agrupaciones sobre la columna de set.
inventory_quantity = inventory_sets.get(['set_num','quantity']).groupby('set_num', as_index = False).sum()
inventory_quantity.head()
```

```python
inventory_quantity.shape
```

```python
inventory_quantity['set_num'].nunique()
```

```python
pd.merge(sets, inventory_quantity, on = 'set_num')
```

```python
# Realizando validaciones de los Merge.
sets.head()
```

```python
inventory_sets.head()
```

```python
inventory_quantity.head()
```

```python
# Realizando validacion de llave unica.
pd.merge(sets, inventory_quantity, on = 'set_num', validate='1:1')
```

```python
# Realizando validacion con el primer set, donde no se tenia llave unica, se genera error.
pd.merge(sets, inventory_sets, on = 'set_num', validate='1:1')
```

```python
# Realizando validacion de uno a muchos.
pd.merge(sets, inventory_sets, on = 'set_num', validate='1:m')
```
