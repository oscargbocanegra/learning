### Combinando Dataframes con la funcion concat.

```python
sets = pd.read_csv(path_input_lego+'star-wars-sets.csv')
sets_2021 = pd.read_csv(path_input_lego+'star-wars-sets_2021.csv')
```

```python
sets.head()
```

```python
sets_2021.head()
```

```python
# Concatenando datasets.
sets_concat = pd.concat([sets, sets_2021], ignore_index=True)
```

## Ejercicio ordenando archivos en un solo dataset.

```python
import os
import pandas as pd
```

```python
# Consultando arhivos locales en ruta datasets / inputs
path_input_lego = "../datasets/input/lego-database/sets-year/"
path_output = "../datasets/output/"
os.listdir(path_input_lego)
#os.mkdir("../datasets/output")
```

```python
# Ordenando listado.
files = sorted(os.listdir(path_input_lego))
# Concatenando datasets lego con la funcion append.
df_lego_sets = pd.DataFrame()

for file in files:
    lego_year = pd.read_csv(path_input_lego+file)
    df_lego_sets = df_lego_sets._append(lego_year, ignore_index = True)
```

```python
df_lego_sets
```

```python
# Concatenando datasets lego con la funcion concat.
files = sorted(os.listdir(path_input_lego))
dfs = [pd.read_csv(path_input_lego+file) for file in files]
# Concatenando datasets lego con la funcion concat.
df_lego_concat = pd.concat(dfs, ignore_index=True)
```

```python
df_lego_concat
```

```python
# Exportando el dataset final al disco local.
df_lego_concat.to_csv(path_output+'lego_sets.csv', index=False)
```

```python
# Consultando arhivos locales en ruta datasets / inputs
path_input_lego = "../datasets/input/lego-database/"
path_output = "../datasets/output/"
os.listdir(path_input_lego)
```
