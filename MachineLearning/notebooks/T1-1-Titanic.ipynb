## Carga de datos a trav√©s de la funci√≥n OPEN

- Recomendable al abrir un fichero en modo "w" (escritura) har√° que dicho fichero pierda su contenido (en caso de tenerlo y querer conservarlo). Si se quiere agregar contenido a este es recomendable abrirlo en modo "a" (append), lo cual posicionar√° el cursor al final y no borrar√° el contenido del mismo.

```python
# Cargando archivos gigantes con funcion OPEN en modo lectura "r"
data4 = open(mainpath +"/"+ "customer-churn-model/Customer Churn Model.txt","r")
```

```python
# Extraer informacion linea a linea.
cols = data4.readline().strip().split(",")
n_cols = len(cols)
```

```python
# Contador para buscar el numemro de filas.

counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
# Rellenar el diccionario con el for.

for line in data4:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter +=1

print("El data set tiene %d filas y %d columnas" % (counter,n_cols))
```

```python
# Creando un DataFrame a partir del diccionario del punto anterior.
df3 = pd.DataFrame(main_dict)
df3.head()
```

## Lectura y escritura de ficehros.

```python
infile = mainpath +"/"+ "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath +"/"+ "customer-churn-model/Tab2 Customer Churn Model.txt"
```

```python
# Escribiendo en Tab2 con el nuevo formato de separacion "\t"
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

```python
df4 = pd.read_csv(outfile, sep="\t")
df4.head()
```
