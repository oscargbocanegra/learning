## Ejemplos de parametros funcion read_csv.

- Otra opcion para cargar datos.
- usando dtype={"columna_a":np.float64, "columna_b":np.int32}
- header: Especifica que fila es la cabecera.
- names: indica el nombre de las columnas. names={"columna_a", "columna_b"}
- skiprows: para saltar un numero determinado de filas. skiprows=12
- index_col: No se aplican etiquetas a cada una de las filas. index_col=None
- skip_blank_lines: Booleano (True o False) las lineas en blanco seran saltadas por NaN.
