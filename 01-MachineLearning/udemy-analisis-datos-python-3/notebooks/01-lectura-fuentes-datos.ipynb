# 01-Creacion de DataFrames.
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html

```python
import pandas as pd
import os
```

```python
pokemon = ["Bulbasaur","Squirtle","Charizard","Pikachu","Onix","Geodude","Vulpix","Starmie"]
tipos = ["Grass / Poison","Water","Fire / Flying","Electric","Rock / Ground","Rock / Ground","Fire","Water"]
numero_pokedex = [1,4,7,25,95,74,37,121]
```

```python
dict_pokemon = {
    'Nombre' : pokemon,
    'Tipo' : tipos,
    'Pokedex' : numero_pokedex
}
```

```python
dict_pokemon
```

```python
# Asignando a una variable dataframe, la lista
df_pokemon = pd.DataFrame(dict_pokemon)
```

```python
# Consultanddo el dataframe.
df_pokemon
```

```python
# Accediendo al nombre de las columnas.
df_pokemon.columns
```

```python
# Accediendo al rango de las filas.
df_pokemon.index
```

```python
df_pokemon.shape
```

```python
# Accediendo al contenido del dataframe, sin columndas ni indices.
df_pokemon.values
```

```python
# Acceder a una columna en particular.
df_pokemon['Nombre']
```

```python
# Acceder a una columna en particular con notacion de punto.
df_pokemon.Nombre
```

```python
# Acceder a varias columnas en particular.
df_pokemon[['Nombre','Tipo']]
```

```python
# Acceder a varias columnas en particular con el metodo GET.
df_pokemon.get(['Nombre','Tipo'])
```

### Otras formas de crear  DataFrames.

```python
pokemones = [['Bulbasaur', 'Grass / Poison', 1],
       ['Squirtle', 'Water', 4],
       ['Charizard', 'Fire / Flying', 7],
       ['Pikachu', 'Electric', 25],
       ['Onix', 'Rock / Ground', 95],
       ['Geodude', 'Rock / Ground', 74],
       ['Vulpix', 'Fire', 37],
       ['Starmie', 'Water', 121]]
```

```python
df_pokemones = pd.DataFrame(pokemones)
df_pokemones
```

```python
# Asignando columndas al dataframe.
df_pokemones.columns = ["Nombre", "Tipo", "Pokedex"]
df_pokemones
```

```python
# Creando dataframe y asignando columnas en una sola linea.
df_pokemones2 = pd.DataFrame(df_pokemones, columns = ["Nombre", "Tipo", "Pokedex"])
df_pokemones2
```
